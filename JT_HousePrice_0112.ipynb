# 分群分析
[距離計算方式](http://www.cnblogs.com/heaad/archive/2011/03/08/1977733.html)

### 載入分群函數 `AgglomerativeClustering`以及`numpy`模塊

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
```

使用`AgglomerativeClustering`進行分群，並指定分群數量`n_cluster`為$3$

```python
cluster = AgglomerativeClustering(n_clusters = 3)
```

* 使用`np.reshape`是為了符合分群函數`AgglomerativeClustering`的需求，並將結果儲存為`temp`
* 使用上面的分群`cluster`搭配`fit_predict`將`temp`分為$3$群
* 最後得到的結果如下，每一筆`house_price.target`會被分為`{0,1,2}`其中一群

```python
temp = np.reshape(a = house_price.target,newshape = [-1,1])
print(house_price.target.shape, temp.shape)
cluster.fit_predict(temp)
```

* 分群結果會儲存在`cluster.labels_`，這裡將分群結果作為一個新欄位`clustering`新增在`df`

```python
df['clustering'] = cluster.labels_
```

```python
df.head()
```

# 對數轉換

複製原始資料集`df`為`df_new`用於對數轉換，以便於之後比較對數轉換前後的散佈圖與模型表現差異

```python
df_new = df.copy() 
```

$y=ab^x \rightarrow \log y=\log a + x\log b$

對數轉換可以使變數的指數關係轉換為線性關係，讓迴歸模型表現更好

```python
df_new['CRIM'] = np.log(df_new['CRIM'])
df_new['LSTAT'] = np.log(df_new['LSTAT'])
```

* 上排為`df`原始資料集還沒對數轉換的`CRIM`與`LSTAT`
* 下排為`df_new`經過對數轉換後的`CRIM`與`LSTAT`

可以發現確實有從指數關係轉為線性關係。

```python
sns.lmplot(x='CRIM',y='price',data=df,fit_reg=False,truncate=False) # df
sns.lmplot(x='CRIM',y='price',data=df_new,fit_reg=False,truncate=False) # df_new
sns.lmplot(x='LSTAT',y='price',data=df,fit_reg=False,truncate=False) # df
sns.lmplot(x='LSTAT',y='price',data=df_new,fit_reg=False,truncate=False) # df_new
```

# 建模 - 線性迴歸模型

這裡要比較對數轉換前後的差異

`LinearRegression`：最基礎的線性迴歸模型，目的在於預測房價的多寡。

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression() 
```

### 對數轉換前`df`

* `lm.fit`：`fit`訓練模型函數
* `df.iloc[:,:-1]`：挑出除了最後一個欄位的所有資料
* `df.iloc[:,-1]`：挑出最後一個欄位的所有資料

```python
df.iloc[:,-1]
```

```python
lm.fit(X = df.iloc[:,:-1], y = df.iloc[:,-1])
```

`lm.score`：計算方式是使用`df.iloc[:,:-1]`預測，並與目標房價`df.iloc[:,-1]`做比較，得到分數。

$R-square$模型評價分數，分數越高越好，值域為$(0,1]\$，此分數表示模型中有多少`variance`可以被特徵所解釋。

這裡分數約為$0.74$，表示特徵可以在此模型解釋百分之$74\%$的`variance`。

```python
lm.score(df.iloc[:,:-1], df.iloc[:,-1])
```

### 對數轉換後`df_new`

* `lm_new.score`的分數約為$0.784$，相較於對數轉換前提升約$0.044$，這表示對數轉換對於迴歸模型確實有效果。

```python
lm_new = LinearRegression()
```

```python
lm_new.fit(X = df_new.iloc[:,:-1], y = df_new.iloc[:,-1])
```

```python
lm_new.score(df_new.iloc[:,:-1],df_new.iloc[:,-1])
```
